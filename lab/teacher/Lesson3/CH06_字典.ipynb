## 6.1 字典基本操作

### 1. 範例實作：血型個性查詢
#### 不同血型的人具有不同的個性：設計程式建立 4 筆字典資料：「鍵」為血型，「值」為個性。使用者輸入血型後，若血型存在，就顯示該血型的個性，如果血型不存在，則顯示沒有該血型的訊息。(<dictget.py>)

````python
dict1 = {"A":"內向穩重", "B":"外向樂觀", "O":"堅強自信", "AB":"聰明自然"}
name = input("輸入要查詢的血型:")
blood = dict1.get(name)
if blood == None:  
    print("沒有「" + name + "」血型！")
else:  
    print(name + " 血型的個性為：" + str(dict1[name]))
````

## 6.2 字典進階操作

### 3. 範例實作：輸入及查詢學生成績
#### 為了解學習狀況，老師需要查詢學生成績。建立 3 筆字典資料：「鍵」為學生姓名，「值」為學生成績。老師輸入學生姓名後，若學生姓名存在，就顯示該學生成績，否則就讓老師輸入成績，並將學生資料加入字典。(<in.py>)

````python
dict1 = {"林小明":85, "曾山水":93, "鄭美麗":67}
name = input("輸入學生姓名：")
if name in dict1:  
    print(name + "的成績為 " + str(dict1[name]))
else:  
    score = input("輸入學生分數：")
    dict1[name] = score
    print("字典內容：" + str(dict1))
````

### 5. 範例實作：keys 及 values 顯示世大運獎牌數
#### 台灣主辦世界大學運動會，成績輝煌。請建立 3 筆字典資料：「鍵」是獎牌名稱，「值」為獎牌數，使用 keys 及 values 方法顯示各種獎牌數。(<keyvalue.py>)

```python
-----------------------
得到的 金牌 數目為 26 面
得到的 銀牌 數目為 34 面
得到的 銅牌 數目為 30 面
-----------------------
```

````python
dict1={"金牌":26, "銀牌":34, "銅牌":30}
listkey = list(dict1.keys())
listvalue = list(dict1.values())
for i in range(len(listkey)):
    print("得到的 %s 數目為 %d 面" % (listkey[i], listvalue[i]))
````

### 第五題
#### 5. 目前美金、日幣和人民幣對台幣的兌換匯率分別是「美金:28.02」、「日幣:0.2513」、「人民幣:4.24」。請設計此匯率兌換程式，輸入台幣金額後計算可以兌換多少的美金、日幣和人民幣。

````python
rate = {'USD':28.02, 'JPY':0.2513, 'CNY':4.24}
TWD = float(input("請輸入台幣："))
print("台幣{:.2f}元等於美金{:.2f}元, 日幣{:.2f}元, 人民幣{:.2f}元".format(TWD, TWD/rate['USD'], TWD/rate['JPY'], TWD/rate['CNY']))
````
